 Create a new column named “age_interval”

```
df1.loc[ (df1.Age < 25), 'age_interval'] = "< 25"
df1.loc[ (df1.Age >25)&(df1.Age <35), 'age_interval'] =  "25-35"
df1.loc[ (df1.Age >36)&(df1.Age <50), 'age_interval'] = "36-50"
df1.loc[ (df1.Age > 50), 'age_interval'] = ">50"
df1.head()
```

```
pd.pivot_table(df1, values = ['Survived'], index = 'age_interval', aggfunc = np.sum)
```
